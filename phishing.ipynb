### Imports and dataset reading

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# https://github.com/GregaVrbancic/Phishing-Dataset
data = pd.read_csv('dataset_full.csv')
```

### Training/testing splits

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, random_state=42)
```

## Support Vector Classifiers

#### Polynomial SVC

```python
from sklearn.svm import SVC

Poly = make_pipeline(StandardScaler(), SVC(kernel='poly', random_state=42))
Poly.fit(X_train, y_train);
```

#### Linear SVC

```python
Linear = make_pipeline(StandardScaler(), SVC(kernel='linear', random_state=42))
Linear.fit(X_train, y_train);
```

#### Radial Basis Function SVC

```python
RBF = make_pipeline(StandardScaler(), SVC(kernel='rbf', random_state=42))
RBF.fit(X_train, y_train);
```

#### Sigmoid SVC

```python
Sigmoid = make_pipeline(StandardScaler(), SVC(kernel='sigmoid', random_state=42))
Sigmoid.fit(X_train, y_train);
```

##### Predictions

```python
y_pred_poly = Poly.predict(X_test)
y_pred_linear = Linear.predict(X_test)
y_pred_rbf = RBF.predict(X_test)
y_pred_sigmoid = Sigmoid.predict(X_test)
```

#### Polynomial SVC results

```python
accuracy_poly = accuracy_score(y_test, y_pred_poly)
classificationReport_poly = classification_report(y_test, y_pred_poly)
confusionMatrix_poly = confusion_matrix(y_test, y_pred_poly)

print(f"Accuracy: {accuracy_poly}\n")
print("Classification Report:")
print(classificationReport_poly)
print("Confusion Matrix:")
print(confusionMatrix_poly)
```

#### Linear SVC results

```python
accuracy_linear = accuracy_score(y_test, y_pred_linear)
classificationReport_linear = classification_report(y_test, y_pred_linear)
confusionMatrix_linear = confusion_matrix(y_test, y_pred_linear)

print(f"Accuracy: {accuracy_linear}\n")
print("Classification Report:")
print(classificationReport_linear)
print("Confusion Matrix:")
print(confusionMatrix_linear)
```

#### Radial Basis Function SVC results

```python
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
classificationReport_rbf = classification_report(y_test, y_pred_rbf)
confusionMatrix_rbf = confusion_matrix(y_test, y_pred_rbf)

print(f"Accuracy: {accuracy_rbf}\n")
print("Classification Report:")
print(classificationReport_rbf)
print("Confusion Matrix:")
print(confusionMatrix_rbf)
```

#### Sigmoid SVC results

```python
accuracy_sigmoid = accuracy_score(y_test, y_pred_sigmoid)
classificationReport_sigmoid = classification_report(y_test, y_pred_sigmoid)
confusionMatrix_sigmoid = confusion_matrix(y_test, y_pred_sigmoid)

print(f"Accuracy: {accuracy_sigmoid}\n")
print("Classification Report:")
print(classificationReport_sigmoid)
print("Confusion Matrix:")
print(confusionMatrix_sigmoid)
```

## Random Forest Classifier + results

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(n_estimators=100, random_state=42)
RFC.fit(X_train, y_train)
y_pred_RFC = RFC.predict(X_test)

accuracy_RFC = accuracy_score(y_test, y_pred_RFC)
classificationReport_RFC = classification_report(y_test, y_pred_RFC)
confusionMatrix_RFC = confusion_matrix(y_test, y_pred_RFC)

print(f"Accuracy: {accuracy_RFC}\n")
print("Classification Report:")
print(classificationReport_RFC)
print("Confusion Matrix:")
print(confusionMatrix_RFC)
```

## Gaussian Naive Bayes Classifier + results

```python
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB()
GNB.fit(X_train, y_train)
y_pred_GNB = GNB.predict(X_test)

accuracy_GNB = accuracy_score(y_test, y_pred_GNB)
classificationReport_GNB = classification_report(y_test, y_pred_GNB)
confusionMatrix_GNB = confusion_matrix(y_test, y_pred_GNB)

print(f"Accuracy: {accuracy_GNB}\n")
print("Classification Report:")
print(classificationReport_GNB)
print("Confusion Matrix:")
print(confusionMatrix_GNB)
```

## Bernoulli Naive Bayes Classifier + results

```python
from sklearn.naive_bayes import BernoulliNB

BNB = BernoulliNB()
BNB.fit(X_train, y_train)
y_pred_BNB = BNB.predict(X_test)

accuracy_BNB = accuracy_score(y_test, y_pred_BNB)
classificationReport_BNB = classification_report(y_test, y_pred_BNB)
confusionMatrix_BNB = confusion_matrix(y_test, y_pred_BNB)

print(f"Accuracy: {accuracy_BNB}\n")
print("Classification Report:")
print(classificationReport_BNB)
print("Confusion Matrix:")
print(confusionMatrix_BNB)
```

## k-Nearest Neighbors Classifier + results

```python
from sklearn.neighbors import KNeighborsClassifier

kNN = KNeighborsClassifier(n_neighbors=3)
kNN.fit(X_train, y_train)
y_pred_kNN = kNN.predict(X_test)

accuracy_kNN = accuracy_score(y_test, y_pred_kNN)
classificationReport_kNN = classification_report(y_test, y_pred_kNN)
confusionMatrix_kNN = confusion_matrix(y_test, y_pred_kNN)

print(f"Accuracy: {accuracy_kNN}\n")
print("Classification Report:")
print(classificationReport_kNN)
print("Confusion Matrix:")
print(confusionMatrix_kNN)
```

## Logistic Regression Classifier + results

```python
from sklearn.linear_model import LogisticRegression

LREG = make_pipeline(StandardScaler(), LogisticRegression(max_iter=10000, random_state=42))
LREG.fit(X_train, y_train)
y_pred_LREG = LREG.predict(X_test)

accuracy_LREG = accuracy_score(y_test, y_pred_LREG)
classificationReport_LREG = classification_report(y_test, y_pred_LREG)
confusionMatrix_LREG = confusion_matrix(y_test, y_pred_LREG)

print(f"Accuracy: {accuracy_LREG}\n")
print("Classification Report:")
print(classificationReport_LREG)
print("Confusion Matrix:")
print(confusionMatrix_LREG)
```

##### Dummy classifier for graphs

```python
from sklearn.dummy import DummyClassifier
Dummy = DummyClassifier(strategy='uniform', random_state=42)
Dummy.fit(X_train, y_train);
```

## Receiver Operating Characteristic Curves

```python
from sklearn.metrics import RocCurveDisplay
import matplotlib.pyplot as plt

ROC_curves = plt.subplot()
RocCurveDisplay.from_estimator(Poly, X_test, y_test, ax=ROC_curves, name="Poly SVC")
RocCurveDisplay.from_estimator(Linear, X_test, y_test, ax=ROC_curves, name="Linear SVC")
RocCurveDisplay.from_estimator(RBF, X_test, y_test, ax=ROC_curves, name="RBF SVC")
RocCurveDisplay.from_estimator(Sigmoid, X_test, y_test, ax=ROC_curves, name="Sigmoid SVC")
RocCurveDisplay.from_estimator(RFC, X_test, y_test, ax=ROC_curves, name="RFC")
RocCurveDisplay.from_estimator(GNB, X_test, y_test, ax=ROC_curves, name="GNB")
RocCurveDisplay.from_estimator(BNB, X_test, y_test, ax=ROC_curves, name="BNB")
RocCurveDisplay.from_estimator(kNN, X_test, y_test, ax=ROC_curves, name="kNN")
RocCurveDisplay.from_estimator(LREG, X_test, y_test, ax=ROC_curves, name="Logistic Regression")
RocCurveDisplay.from_estimator(Dummy, X_test, y_test, ax=ROC_curves, name="Random Classifier")

ROC_curves.set_xlabel("False Positive Rate")
ROC_curves.set_ylabel("True Positive Rate")
ROC_curves.set_title("ROC Curves")
ROC_curves.legend();
```

## Precision Recall Curves

```python
from sklearn.metrics import PrecisionRecallDisplay

PRC_curves = plt.subplot()
PrecisionRecallDisplay.from_estimator(Poly, X_test, y_test, ax=PRC_curves, name="Poly")
PrecisionRecallDisplay.from_estimator(Linear, X_test, y_test, ax=PRC_curves, name="Linear")
PrecisionRecallDisplay.from_estimator(RBF, X_test, y_test, ax=PRC_curves, name="RBF")
PrecisionRecallDisplay.from_estimator(Sigmoid, X_test, y_test, ax=PRC_curves, name="Sigmoid")
PrecisionRecallDisplay.from_estimator(RFC, X_test, y_test, ax=PRC_curves, name="RFC")
PrecisionRecallDisplay.from_estimator(GNB, X_test, y_test, ax=PRC_curves, name="GNB")
PrecisionRecallDisplay.from_estimator(BNB, X_test, y_test, ax=PRC_curves, name="BNB")
PrecisionRecallDisplay.from_estimator(kNN, X_test, y_test, ax=PRC_curves, name="kNN")
PrecisionRecallDisplay.from_estimator(LREG, X_test, y_test, ax=PRC_curves, name="Logistic Regression")
PrecisionRecallDisplay.from_estimator(Dummy, X_test, y_test, ax=PRC_curves, name="Random Classifier")

PRC_curves.set_xlabel("Precision")
PRC_curves.set_ylabel("Recall")
PRC_curves.set_title("Precision-Recall Curves")
PRC_curves.legend();
```
